# SNT - La photo numérique

## 2. Formats des images
#### 2.1 Introduction - <a href="http://lycee.Lagrave.free.fr/isn/numerisation/2-traitement_numerique_de_l'image.pdf" target="_blank">traitement numerique de l'image</a>


<div class="alert alert-danger">
<p>Pour un ordinateur, <b>une image est un ensemble de pixels arrangés sous la forme d'un tableau à deux dimensions</b>.</p>
<ul> <li> Une image dite <b>en noir et blanc</b> est composée de pixels noirs ou blancs :<br/> les deux valeurs possibles des pixels sont $0$ pour le noir et $1$ pour le blanc.</lio>

<li>Une image <b>en niveaux de gris</b> est composée de pixels de valeurs numériques représentant l'intensité de la luminosité,<br/> généralement les valeurs sont entières entre $0$ (pour le noir) et $255$ (pour le blanc).</li>

<li>Une image <b>en couleurs</b> est composée de pixels représentés par trois valeurs : celle représentant l'intensité du rouge (R), celle représentant l'intensité du vert (V) et celle représentant l'intensité du bleu (B). Chaque valeur est comprise entre $0$ et $255$.</li>
</ul>
</div>



* **Remarque ** On parle ici d’un codage sur huit bits : en effet, chaque nombre compris entre $0$ et $255$ peut être associé à un « mot » de 8 bits   
  (un tel mot est par exemple : $10010100$ qui code la nuance $148$).
    Dans la suite, on parlera d'octet, 1 octet = 8 bits.*


####  2.2 Les formats sans compression
Formats portables

<div class="alert alert-danger">
    <p>Les formats de fichiers d'images $\texttt{PBM}$ (portable bitmap), $\texttt{PGM}$ (portable grayscalemap) et $\texttt{PPM}$ (portable pixmap), offrent une solution simple à tout programmeur confronté au problème de la manipulation de fichiers d'images.</p>

<p>Ces formats n'effectuent pas de compression, c'est-à-dire que tous les pixels sont présents dans l'image.</p> <p>L'avantage de ces formats est la qualité des images car sans compression il n'y a pas de perte de qualité. En revanche cela donne des fichiers très volumineux car les pixels sont représentés par des caractères $\texttt{ASCII}$ qui occupent 1 octet chacun.</p>
</div>

Ces fichiers, avec les trois formats, sont construits sur la même base :

* Le nombre « magique » du format (P1 pour les images en noir et blanc, P2 pour celles en niveaux de gris et P3 pour les images en couleurs).
* Un caractère d'espacement qui peut être un espace, une tabulation ou un retour à la ligne.
* La largeur de l'image (valeur décimale, codée en $\texttt{ASCII}$) suivie d'un caractère d'espacement, la longueur de l'image (valeur décimale, $\texttt{ASCII}$) suivie d'un caractère d'espacement.
* Uniquement pour $\texttt{PGM}$ et $\texttt{PPM}$ : l'intensité maximum (valeur décimale comprise entre $0$ et $255$, codée en $\texttt{ASCII}$) suivie d'un caractère d'espacement.
* Un caractère d'espacement qui peut être un espace, une tabulation ou une nouvelle ligne.

Les valeurs des pixels de l'image :
* sont données ligne par ligne en partant du haut.
* chaque ligne est codée de gauche à droite et contient moins de 70 caractères. De plus, toutes les lignes commençant par # sont ignorées.

En général, il est immédiatement modifié et enregistré dans un autre format.<br/> Il existe un autre non compressé, le format <b>Tiff</b> qui est adapté à l'impression.</p>

#### 2.3 Formats avec compression

Il y a deux grandes familles : les méthodes de compression d'image sans perte et celles avec perte.

##### Formats sans perte
<div class="alert alert-danger"><p>L'une des méthodes les plus importantes de compression d'image sans perte est la <b>méthode du codage des répétitions</b> <br/>(comme LRE "run-length encoding")</p>

<p>Cet encodage consiste, pour chaque suite de pixels de la même couleur, à coder uniquement le nombre de pixels puis la couleur de la séquence. Ainsi sur l'image noir et blanc représentée par la suite de pixels blancs (W) ou noirs (B) suivante :<br/>
$\texttt{WWWWWWWBWWWWWWWWWWWBBBBWWWWWWWWWWWWWWWWWWWWBWWWWWWWWW}$ le codage est : $\texttt{7W1B11W4B20W1B9W}$
</p>
<p>En général le codage comporte moins de caractères (il y a bien compression). Mais ce n'est pas toujours le cas comme le montre l'image :<br/>
$\texttt{WBWBWBWBWB}$ qui se code par $\texttt{1W1B1W1B1W1B11W1B1W1B}$</p>
</div>

##### Formats avec perte

<div class="alert alert-danger"><p>Il existe de nombreuses techniques de compression avec perte. Certains algorithmes de compression utilisent le fait que l'oeil humain est assez sensible à la <b>luminance</b> (la luminosité) mais peu à la <b>chrominance</b> (la teinte) d'une image.<br/> Ainsi <b>les pixels ayant des teintes proches sont uniformisés</b>.</p>
    <p>Les couleurs sont donc modifiées afin d'être uniformes dans les zones ayant des teintes proches (ce que l'oeil humain a du mal à distinguer).<br/> Le processus de compression est irréversible, puisqu'il y a perte de données.</p>
</div>
